# Reversing a List in Python

<h3>1. Using List Slicing to Reverse an Array in Python</h3>

```python
arr = [11,12,13,14,15]
print("Printing original Array", arr)

res = arr[::-1]  #reversing using List slicing
print("Resultant new reversed array",res)
```

<h3>2. Using reverse() Method</h3>

```python
arr = [11, 22, 33, 44, 55]
print("Before reversal array is:", arr)

arr.reverse()
print("After reversing array", arr)
```

<h3>3. Using reversed() Method</h3>

```python
arr = [100,102,103,104,105]
print("Original array is:", arr)

result=list(reversed(arr))
print("Resulatant new reversed Array" ,result)
```
